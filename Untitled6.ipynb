# Expresión analítica para la formula de representación de Enneper 




$\frac{1}{2}(1-\nu^2(z))μ(z))$

```python
A_1=1/2*(1-A*A)*B
C=simplify(A_1)
C
```

$\frac{i}{2}(1+ν^2(z))μ(z)$

```python
D=I/2*(1+A*A)*B
D=simplify(D)
D
```

 $μ(z) ν(z)$

```python
E=A*B
E
```

$∫\frac{1}{2}(1-\nu^2(z))μ(z))dz$

```python
F=integrate(C,z)
F=simplify(F)
F
```

$∫\frac{i}{2}(1+\nu^2(z))μ(z))dz$





```python
G=integrate(D,z)
G=simplify(G)
G
```

$∫\nu(z)μ(z))dz$

```python
H=integrate(E,z)
H=simplify(H)
H
```

Por  lo tanto la superficie minima asociada a la formula de representacion de Enneper esta caracterizada por la parte real de la siguiente exprersión 

```python
J=[F,G,H]
J
```

# Propiedades geometricas 

# Curvatura de Gauss

Como la curvatura de gaus para una superficie parametrizada por la formula de Enneper-Weierstrass, esta dada por 


\begin{equation*}
\begin{split} 
 K(w)= -\left [  \frac{4⋅\left | v '(w)\right |}{\left | \mu(w) \right |(1+\left | v(w) \right |^2)^2} \right ]^2
\end{split}
\end{equation*}

Entonces para el caso anterior vamos a tener que 

 \begin{equation*}
\begin{split} 
 K(w)&= -\left [  \frac{4⋅\left | 0.5(c-k)⋅e^{b+0.5(c-k)z}\right |}{\left | e^{a+k⋅z} \right |(1+\left | e^{b+0.5(c-k)z} \right |^2)^2} \right ]^2 \\
 &=-\left [  \frac{2(c-k)⋅ e^{a_2+0.5(c-k)u}}{ e^{a_1+k⋅u}⋅ (1+ e^{2⋅ a_2+(c-k)u})^2} \right ]^2 \\
 &=-\frac{4(c-k)^2 \cdot e^{2\cdot(a_1-a_2)}⋅e^{(c-3k)\cdot u}}{(1+ e^{2⋅ a_2}⋅e^{(c-k)u})^4}
\end{split}
\end{equation*}

# Ecuación de las lineas de curvatura  y lineas asintoticas 

Por otra parte sabemos que las lineas de curvatura y las lineas asintoticas  de una superficie que se pueda representar por medio de la formula de weiertrass estan caracterizadas por las ecuaciónes 


$$ \Re e \left [ \mu (\gamma ) \cdot \dot{\nu }(\gamma ) \cdot \dot{\gamma }^2 \right ] = 0$$
$$Im \left [ \mu (\gamma ) \cdot v '(\gamma ) \cdot \dot{\gamma }^2 \right ] = 0$$
Lo cual para el caso planteado anteriormente es equivalente a decir 

\begin{equation*}
\begin{split} 
\Re e \left [ \mu (\gamma ) \cdot \dot{\nu }(\gamma ) \cdot \dot{\gamma }^2 \right ]
&=\Re e\left [ e^{a+b+0.5\cdot (c+k)\cdot z(t)}\cdot( \dot{u}(t) +i\dot{v}(t))\right ]\\
&= e^{a_1+a_2}e^{0.5\cdot (c+k)\cdot u(t)}\Re e\left [ e^{i(b_1+b_2)}e^{0.5 i\cdot (c+k)\cdot v(t)} (\dot{u}(t) +i\dot{v}(t))\right ]\\
&= e^{a_1+a_2}e^{0.5\cdot (c+k)\cdot u(t)}\Re e\left [(\cos(b_1+b_2+0.5⋅(c+k)u(t))+i\sin( b_1+b_2+0.5⋅(c+k)v(t)))⋅(\dot{u}(t) +i\dot{v}(t))\right ] \\
&= e^{a_1+a_2}e^{0.5\cdot (c+k)\cdot u(t)}\left ( \cos(b_1+b_2+0.5⋅(c+k)u(t))⋅ \dot{u}(t)-\sin( b_1+b_2+0.5⋅(c+k)v(t)))⋅\dot{v}(t) \right )
\end{split}
\end{equation*}

Asi pues ,vamos a tener que las lineas de curvatura y las lines de curvatura  van a estar dadas por las ecuaciónes 

\begin{equation*}
\begin{split} 
 \cos(b_1+b_2+0.5⋅(c+k)u(t))⋅ \dot{u}(t)-\sin( b_1+b_2+0.5⋅(c+k)v(t))⋅\dot{v}(t)=0\\
\end{split}
\end{equation*}
y
\begin{equation*}
\begin{split}
\sin(b_1+b_2+0.5⋅(c+k)u(t))⋅ \dot{u}(t)+\cos( b_1+b_2+0.5⋅(c+k)v(t))⋅\dot{v}(t)=0
\end{split}
\end{equation*}
