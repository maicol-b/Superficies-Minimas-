```python
D7= diff(expresion4,z,2)
D8= diff(expresion5,z,2)
D9= diff(expresion6,z,2)
[D7,D8,D9]
```

$\alpha'(z) \times \alpha''(z) $

```python
S=D5*D9-D6*D8
S1=D6*D7-D4*D9
S2=D4*D8-D5*D7
S=simplify(S)
S1=simplify(S1)
S2=simplify(S2)
```

```python
[S,S1,S2]
```

$\left \| \alpha'(z) \right \|$

```python
M1=sqrt(D7*D7+D8*D8+D9*D9)
M1=simplify(M1)
M1
```

```python
M2=D4*D7+D5*D8+D6*D9
M2=simplify(M2)
M2
```

```python
H=M*M*M1*M1-M2*M2
H=simplify(H)
H
```
